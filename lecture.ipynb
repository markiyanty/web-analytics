First practice: Exploring Central limit theory with numpy
Центральна гранична теорема стверджує, що сума деякої кількості незалежних і однаково розподілених випадкових величин із скінченною дисперсією буде прямувати до нормального розподілу із збільшенням кількості цих величин
Central limit theorem (CLT) states that, under appropriate conditions, the distribution of a normalized version of the sample mean converges to a standard normal distribution. This holds even if the original variables themselves are not normally distributed. There are several versions of the CLT, each applying in the context of different conditions.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
s = []
for i in range(100):
    s.append(random.random())

p = []
for i in range(100):
    p.append(random.random())

np.mean(s)
```

```python
mean_s = []
s = []

for i in range(10000):
    s.append(random.random())
    mean_s.append(np.mean(s))
    
mean_p = []
p = []

for i in range(10000):
    p.append(random.random())
    mean_p.append(np.mean(p))


plt.plot(mean_s, label = 's')
plt.plot(mean_p, label = 'p')
plt.legend()
plt.show()
```

```python
cube = [1,2,3,4,5,6]

obs = []
for i in range(1,1000000):
    obs.append(random.choice(cube))
print(obs)
```

```python
for i in range(100):
    print(obs[i*10:i*10+9])
```

```python
mean_10 = []
for i in range(10000):
    mean_10.append(np.mean(obs[i*10:i*10+9]))
    
mean_10
```

```python
plt.hist(mean_10, bins = 20)
plt.show()
```

```python
np.mean(mean_10)
```
