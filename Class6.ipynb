Z-test (we are assuming that std is known for the population):



H_0: $\mu_m$ == $\mu_f$ \\
H_1: $\mu_m$ != $\mu_f$

$\alpha$ = 0.05

Step 2: \\
Reject $H_0$ in favor of $H_1$ if either \\
\begin{equation}
X̄_m - X̄_f > c \\
or \\
X̄_m - X̄_f < -c \\
\end{equation}
for some c to be determined

Step 3: \\
Find c, if $\alpha$=0.05 \\
\begin{equation}
\alpha = 1 - P(-c < (X̄_m - X̄_f) < c) \\
\alpha = 1 - P( \frac{-c-(μ_m - μ_f)}{\sqrt{\sigma_m^2 / n_m + \sigma_f^2 / n_f}}< \frac{(X̄_m - X̄_f) - (μ_m - μ_f)}{\sqrt{\sigma_m^2 / n_m + \sigma_f^2 / n_f}} < \frac{c+(μ_m - μ_f)}{\sqrt{\sigma_m^2 / n_m + \sigma_f^2 / n_f}})
\end{equation}
